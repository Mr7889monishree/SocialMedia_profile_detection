```python
#separating data and labels , here label is the output which is 'fake' column
x_train = df_train.drop(columns='fake',axis=1)
x_train
```

```python
x_test=df_test.drop(columns='fake',axis=1)
x_test
```

```python
y_train=df_train['fake']
y_train
```

```python
y_test=df_test['fake']
y_test
```

MODEL EVALUATION


```python
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
```

```python
scaler_x = StandardScaler()
X_train = scaler_x.fit_transform(x_train)
X_test=scaler_x.transform(x_test)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
svm_model=SVC()
svm_model.fit(X_train,y_train)
```

```python
y_pred = svm_model.predict(X_test)
accuracy = accuracy_score(y_test,y_pred)
accuracy
```

```python
precision=precision_score(y_test,y_pred,average="binary")
precision
```

```python
recall=recall_score(y_test,y_pred,average="binary")
recall
```

PREDICTIVE SYSTEM

```python
input_data=(0,0.12,1,0.0,0,0,0,1,0,31,213)

#converting our given data into the array
input_data_as_numpy_array = np.asarray(input_data)


#reshaping in the way it goes through all the column in the input data given  inorder to match the expected output from the input data values given
input_data_reshape=input_data_as_numpy_array.reshape(1,-1)

#standardizing the values of the input data given from the training or testing data set for prediction
std_data = scaler_x.transform(input_data_reshape)

print("Standardized Data:",std_data)

#making the prediction on the model of svm that is standardized in the previous step
prediction=svm_model.predict(std_data)

print("Prediction is Given as:",prediction)

#Checking if the given prediction value is fake profile or true profile

if(prediction[0]==0):
  print("This is a True profile!")
else:
  print("This is a Fake Profile")
```
